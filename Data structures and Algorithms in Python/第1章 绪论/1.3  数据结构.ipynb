# <center>数据结构</center>

## 常用术语
数据（Data）：计算机程序能处理的符号形式的总和<br>
数据元素（Data Element）:数据的基本单位，在程序中通常作为一个整体考虑和处理<br>
数据结构（Data Structures）:一组数据元素（结点）按照一定方式构成的复合数据形式,以及作用于这些元素或者结构上的一些函数或操作<br>
逻辑结构：数据元素之间有某种特定的逻辑关系。这是元素之间的抽象关系，与实现无关<br>
物理结构：数据的逻辑结构在计算机存储器中的映射（或表示），又称存储结构，或数据结构的存储表示<br>



## Python数据类型
- 整数型：int型，正负整数
- 浮点数：float型，由整数和小数构成
- 复数：由实部和虚部构成
- 常量：不能改变现有值的变量，如pi或者自然对数e
- 布尔型：True和False
- 字符串：文本的数据类型


## Python中常用的数据结构
- 列表(list)：一种不同数据类型的有序集合

    创建  
        a_list = [elements]
    访问  
        a_list[index]   下标访问,左开右闭
    方法  
        添加到末尾 a_list.append(element)  
        删除末尾元素 a_list.pop() 
        删除指定下标元素 a_list.pop(index)
        插入元素 a_list.inset(index, element)
        按值查找 a_list.index(element)
        移除指定元素 a_list.remove(element)
        删除整个列表 del a_list
        
- 元组(tuple)：一直有序列表，但是元组中的变量是不能改变的

    创建
        a_tuple = (elements)
    基本操作
        删除整个元组 del a_tuple
        
- 集合(set)：一系列无序的，不重复的组合体

    创建
        a_set = {elements}  或者 a_set = set([list])
    基本操作
        不支持下标查找
        intersection(交), difference(差)，union(联合)和sysmmetric difference(对称差集)等数学运算
        x & y # 交集
        x - y # 差集 
        x | y # 并集
        
- 字典(dict):存放无序的键/值(key/value)映射类型数据的容器，其中关键字必须为不可变对象

    创建
        a_dict = {key:value}
    基本操作
        取值 a_dict[key]
        判断关键字是否存在 'key' in a_dict
