# <center>计算机问题求解</center>

## 抽象
使用计算机是为了解决实际问题。牵涉到:

1. 能用计算机解决的问题的性质，特点**(抽象)**
2. 用计算机解决问题的途径和方法**(实现)**

## 建模
解决一个实际问题，就要在计算机里建立该问题的求解模型：

1. 处理实际问题中的各种对象及其相互关系,把这方面信息**映射**到计算机可以处理的表示形式
2. 把实际问题的求解过程**映射**到一个计算过程，用程序实现该过程

<font color=red>从上面这段话可以看出，程序设计/软件开发就是要实现这两个映射。而且实现这两个映射与使用哪种计算机编程语言无关。   
这就是程序设计/软件开发的高度抽象性。</font>


## 程序开发过程

1. 分析阶段：弄清需要求解的问题，给出尽可能严格的描述。这一工作阶段通常被称为"需求分析"。
2. 设计阶段：设计出与实际问题对应的求解方案，进而设计出有关实现的细节方案（信息到数据表示的映射，规划求解过程等。这是数据结构这门课程的核心）
3. 编码阶段：用某种适当的编程语言实现这个模型，做出一个可能由计算机执行的实际计算模型，也就是一个程序。
4. 测试和维护：确认得到的程序能解决问题，以及为满足某些实际目标或需要而修改程序，扩充功能等。
