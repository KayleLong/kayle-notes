### 3.12 删除重复行


使用drop_duplicates方法，但是需要留意其参数！

```python
# 删除重复行，查看输出结果的前两行
dataframe.drop_duplicates().head(2)
```

对比删除前后的结果：

```python
print('Number of rows in the original DataFrame:', len(dataframe))
print('Number of rows in after deduping DataFrame:', len(dataframe.drop_duplicates()))
```

发现上述方案实际上**没有删除任何行**！

这是因为drop_duplicates()方法默认删除那些所有列都完美匹配的行。在这种情况下dataframe中的每一行都是唯一的(默认的索引值即行号是唯一的)。

但是只用某些列作为索引，来检查重复行会是一个很常见的需求。

可以使用**subset**参数

```python
# 用'Sex'作为索引，删除重复行
dataframe.drop_duplicates(subset=['Sex'])
```

默认保留先出现的行。

如果想保留最后的出现的行，可以使用**keep**参数

```python
dataframe.drop_duplicates(subset=['Sex'], keep='last')
```

### 3.13 根据值对行分组

分组操作是非常常用的操作，pandas中通过groupby()方法来进行分组。

```python
# 根据Sex列的值来对行进行分组，并计算每一组的平均值
dataframe.groupby('Sex').mean()
```

```python
dataframe.groupby('Sex')
```

只使用groupby()方法返回的是一个内存对象！

**groupby()方法需要和一些作用于组(group)的操作配合使用。**

```python
dataframe.groupby('Survived').count()
```

这个表格不符合我们的预期

```python
# 按行分组，计算行数
dataframe.groupby('Survived')['Name'].count()
```

注意到 Name 被加到 groupby()方法**后面**了。

这是因为特定的描述性统计量只对特定的数据类型才有分组的意义。

例如：按性别来计算平均年龄。

上面的例子，是先按乘客是否幸存做了分组，再按 Name（Name是唯一值，换成PassengerId也是可以的）计算数量。

```python
dataframe.groupby('Survived')['PassengerId'].count()
```

**多重分组**

```python
# 先按照性别分组(1*2)，再按照是否幸存分组(1*2*2)，最后计算平均值
dataframe.groupby(['Sex', 'Survived'])['Age'].mean()
```

### 3.19 连接多个数据帧
使用concatenate()方法来连接(**黏在一起**)多个数据帧。

```python
# 创建数据帧
data_a = {
    'id':['1', '2', '3'],
    'first':['Alex', 'Amy', 'Allen'],
    'last':['Anderson', 'Ackerman', 'Ali']
}
dataframe_a = pd.DataFrame(data_a, columns=['id', 'first', 'last'])

data_b = {
    'id':['4', '5', '6'],
    'first':['Billy', 'Brian', 'Bran'],
    'last':['Bonder', 'Black', 'Balwner']
}
dataframe_b = pd.DataFrame(data_b, columns=['id', 'first', 'last'])

# 沿着行的方向，连接两个数据帧
pd.concat([dataframe_a, dataframe_b], axis=0)
```

使用 axis=1 可以在列的方向上做连接

```python
pd.concat([dataframe_a, dataframe_b], axis=1)
```

### 3.19 合并两个数据帧
**合并**分为：
- 等值连接(inner join)，可以理解为**交**，即合并只有相同内容的列
- 外连接(outer join)，可以理解为**并**，即合并全部内容

使用merge()方法进行数据帧的合并，默认是等值连接。用 on 参数来指定合并哪些列。

```python
employee_data = {
    'employee_id':['1', '2', '3', '4'],
    'name':['Amy Jones', 'Allen Keys', 'Alices Bees', 'Tim Horton']
}
dataframe_employees = pd.DataFrame(employee_data, columns=['employee_id', 'name'])

sales_data = {
    'employee_id':['3', '4', '5', '6'],
    'total_sales':[2000, 3000, 3000, 4000]
}
dataframe_sales = pd.DataFrame(sales_data, columns=['employee_id', 'total_sales'])

# 合并数据帧
pd.merge(dataframe_employees, dataframe_sales, on='employee_id')
```

外连接可以通过how参数来指定：

```python
pd.merge(dataframe_employees, dataframe_sales, on='employee_id', how='outer')
```
