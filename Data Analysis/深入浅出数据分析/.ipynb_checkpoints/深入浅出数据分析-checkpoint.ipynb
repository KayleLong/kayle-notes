# <center>深入浅出数据分析</center>

### 目的：
利用Excel等程序组织数据，在R程序中进行进一步的整理，通过散点图和直方图找出有意义的模式，借助启发式算法作为结论，通过实验和假定预测预见未来，再以清楚直观的图形展示分析结果。

### 应知应会的技术：
数据分析的基本步骤、实验方法、最优化方法、假设检验方法、贝叶斯统计方法、主观概率法、启发法、直观图、回归法、误差处理、相关数据库、数据整理技巧等。

### 目录和简介
1. 数据分析引言：分解数据<br>
    如何处理所有的数据材料，如何将原始数据转变成推进现实工作的妙策，如何分解和构建复杂的问题和数据集，进而牢牢把握工作中的各种问题的要害

2. 实验：检验你的理论<br>
一个好实验往往能让你摆脱对观察数据的无限依赖，能帮助你理清因果联系；可靠的实证数据将让你的分析判断更有说服力。

3. 最优化：寻找最大值<br>
有一种数据分析工具能够帮助我们调整决策变量，找出解决方案和优化点，使我们最大限度地达到目标。

4. 数据图形化：图形让你更精明<br>
数据庞杂晦涩，无尽的电子表格令人厌倦不堪，而且确实浪费时间。一幅用纸不多、栩栩如生的清晰图像，却能让你摆脱"一叶障目，不见泰山"的烦恼。

5. 假设检验：假设并非如此<br>
人们需要用庞杂多变的数据预测未来。通过数据分析的仔细推理，分析师能够异常细致地评估大量备选答案，然后将手头的一切信息整合到各种模型中。

6. 贝叶斯统计：穿越第一关<br>
数据收集工作永不停息。必须确保每一个分析过程都充分利用所搜集到的与问题有关的数据。贝叶斯规则，这个规则能帮助你利用基础槪率和波动数据做到明察秋奄。

7. 主观概率：信念数字化<br>
如何利用标准偏差评估数据分布

8. 启发法：凭人类的天性作分析<br>
世事难料，分析并不总是准确有效的。利用既不齐全也不确定的信息，凭经验进行处理，迅速做出决策。奇就奇在这些经验确实能够奏效，因此也是进行 数据分析的重要而必要的工具。

9. 直方图：数字的形状<br>
直方图能够量度数据的分布、差异、集中趋势

10. 回归：预测<br>
回归分析能帮助你预测某些结果值。运用回归分析帮助自己建立模型，预测客户行为。

11. 误差：合理误差<br>
在进行预测的时候指出误差范围。懂得如何控制误差及如何尽量降低误差，从而提高预测可信度

12. 相关数据库：你能关联吗？<br>
如何组织变化多端的多变量数据？

13. 整理数据(数据清洗)：井然有序<br>
乱糟糟的数据毫无用处。许多数据搜集者需要花大量时间整理数据。
