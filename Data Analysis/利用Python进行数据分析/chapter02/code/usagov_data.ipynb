## 3.实现对不同时区进行计数功能

### 普通的方法
在遍历list的过程中将计数值保存到一个字典中。

```python
def get_counts(sequence):
    counts = {}
    for x in sequence:
        if x in counts:
            counts[x] += 1
        else:
            counts[x] = 1
    return counts
```

### 使用Python标准库

事实上，如果了解Python标准库的话，可以使用defaultdict

```python
from collections import defaultdict

def get_counts2(sequence):
    counts = defaultdict(int)# 所有的值初始化为0
    for x in sequence:
        counts[x] += 1
    return counts
```

```python
# 字典对象
counts = get_counts(time_zones)
counts2 = get_counts2(time_zones)
```

## 4.进行排序

### 普通方法
想要得到前10位的时区及其计数值，需要使用到一些有关字典测操作

```python
def top_counts(count_dict, n=10):
    val_key_pairs = [(count, tz) for tz, count in count_dict.items()]# 列表生成器取出count 和 tz
    val_key_pairs.sort()# 排序,默认是升序效果
    return val_key_pairs[-n:]# 切片，取出最后10项数据
```

```python
top_counts(counts)
```

### 假设现在想要按照Windows和非Windows用户对时区统计信息进行分解。
当agent字符串中含有'windows'时，简单的认为是Windows用户。

**数据的清洗**<br>
首先移除缺失agent的数据

```python
cframe = frame[frame.a.notnull()]
```

判断是否为Windows

```python
operating_system = np.where(cframe['a'].str.contains('Windows'), 'Windows', 'Not Windows')
```

```python
operating_system[:5]
```

```python
by_tz_os = cframe.groupby(['tz', operating_system])
```

```python
agg_counts = by_tz_os.size().unstack().fillna(0)
agg_counts[:10]
```

```python
# 用于升序排列
indexer = agg_counts.sum(1).argsort()
indexer[:10]
```

通过take按照升序截取了最后10行

```python
count_subset = agg_counts.take(indexer)[-10:]
count_subset
```

```python
count_subset.plot(kind='barh', stacked=True)
```

由于上面这张图不容易清楚地比较Windows用户的相对比例，因此将各行规范化为'总计为1'

```python
normed_subset = count_subset.div(count_subset.sum(1), axis=0)
normed_subset.plot(kind='barh', stacked=True)
```
