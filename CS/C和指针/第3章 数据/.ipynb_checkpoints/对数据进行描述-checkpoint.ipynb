# <center>数据的描述</center>
程序是对数据进程操作。C语言是一种强类型的编程语言，所以需要从多个角度对于数据进行描述<br>

1. 数据的各种类型
2. 不同类型的数据的特点
3. 如何声明不同类型的数据

# <center>基本数据类型</center>

在C语言中，
<font color=blue size=4>仅有4种基本数据类型——整型、浮点型、指针和聚合类型(如数组和结构等)</font><br>
所有其他的类型都是从这4种基本类型的某种组合派生出来的。


### 整型
整型包括字符char(8位)、短整型short int(16位)、整型int(32位)和长整型long int(64位)。<br>
    
    
整型变量的最小范围

|类型|最小范围|
|--|--|
|char|0到127|
|signed char|-127到127|
|unsigned char|0到255|
|short int|-32767到32767|
|unsigned short int|0到65535|
|int|-32767到32767|
|unsigned int|0到65535|
|long int|-2147483647到2147483647|
|unsigned long int|0到4294967295|


#### 枚举类型
枚举类型的变量实际上是整数类型。

### 浮点型
浮点数包括float、 double 和long double。即单精度、双精度以及拓展精度<br>
浮点数是用一个小数个以及一个以某个假定数为基数的指数组成。例如<br>
.3243F\*16  .1100100010000\* 2²

所以，浮点数不是精确的。因此不可以使用 "==" 来判断两个浮点数的运算结果是否相等。例如<br>
0.1 + 0.6 == 0.7

### 指针类型
指针也是变量，全称指针变量。一般习惯性地称为指针。

变量的值存储于计算机的内存中，每个变量都占据一个特点的位置(由操作系统分配)。每个内存位置都由地址唯一确定并引用。
<font color=blue size=4>指针变量就是一个其值为另一个(一些)内存地址的变量</font>

### 关于字符串
不同于Java、Python，C语言中没有字符串类型。这是因为C语言中的字符串通常存储在字符数组中，可以通过数组名(指针)来访问。而字符串以NUL字节结尾，所以给定存储字符串的首地址就可以依次访问，直到NUL字节，就访问了整个字符串。

# <center>基本类型的声明</center>


### 显式声明
变量声明的基本形式是：
```
说明符(一个或者多个) 声明表达式列表
```
|类型|例子|
|--|--|
|整型|int i=1;|
|浮点型| float i=1.2;|
|数组|int arr[10];|
|指针|int \*a;|


```
int *a;
```
这条语句声明了一个int类型的名字叫a的指针变量。即 a是一个指向int的指针。
<font color=blue size=4>\*操作符执行的是间接访问(间接寻址)操作</font>

### 隐式声明
<font color=red size=4>函数如果不显式地声明返回值的类型，那么函数就默认返回整型</font><br>
隐式声明容易产生歧义，不能够准确表达程序的意图。

### 易错点

指针变量的声明
```
int *a;
int* a;
int *a, b, c;
```

1. 语句2与语句1都是声明一个int类型的指针。但是语句1看上去更为清楚，表达更为准确。
2. 语句3中，a变量为指针变量，b和c是整型变量。


字符串指针
```
char *message = "Hello world";
```
这条语句把message声明为一个指向字符的指针，并用字符串常量中第一个字符的地址对该指针进行初始化。
<font color=red size=4>这个声明看起来是赋给表达式*message，事实上它是赋给message本身的</font>
上面的声明相当于
```
char *message;
message = "Hello world";
```

# <center>变量的三个属性</center>
变量的三个属性——作用域、链接属性和存储类型。这三个属性决定了一个变量的"可视性"(可以在什么地方用)和"生命周期"(变量的值将保持多久)

### 作用域

#### 什么是作用域？
当变量在程序的某个部分被声明时，它只有在程序的一定区域才能被访问。例如：<br>
a文件中的变量，在b文件没有引用a文件时，b文件无法使用a文件中的变量。

编译器可以确认4种不同类型的作用域——文件作用域、函数作用域、代码块作用域和原型作用域(作用范围依次减小)

1. 文件作用域
    任何在代码块之外的声明的标识符都是具有文件作用域。它们可以在声明之处到源文件结尾处都是可以访问的
2. 函数作用域
    函数作用域只适用于goto语句标签
3. 代码块作用域
    位于一对花括号之间的所有语句称为一个代码块，代码块内部的声明都具有代码块作用域
4. 原型作用域
    原型作用域值适用于函数原型中声明的参数名

### 链接属性

一个程序往往都是由多个源文件组成，各个源文件分别被编译之后被链接在一起。<br>
不同源文件中可以使用相同的标识符(变量名)，标识符的链接属性决定如何处理在不同文件中出现的标识符。

链接属性有3种——external（外部）、internal（内部）和none(无)。

1. external 也被称为全局变量(global)。使用 extern 关键字来修改标识符的链接属性
2. internal 局部变量，不可被未被引用的文件访问。使用 static 关键字来修改标识符的链接属性
3. none 普通变量

### 存储类型
变量的存储类型是指存储变量值的内存类型。变量的存储类型决定变量何时创建、何时销毁以及它的值将保持多久。

<font color=blue size=4>计算机中有三个地方可以用于存储变量：普通内存、运行时堆栈、硬件寄存器。</font>

变量的缺省存储类型取决于它的声明位置。

1. 凡是在任何代码块之外声明的变量总是存储于静态内存中，即不属于堆栈内存。这类变量称为静态(static)变量
2. 代码块内部声明的变量存储于堆栈中。这类变量称为自动(automatic)变量
3. 在代码块内部声明的变量加上关键字 static  ，它的存储类型从自动变为静态。
4. 关键字 register 可以用于自动变量的声明，存储在机器的硬件寄存器中。这类变量称为寄存器变量


### 关于 static
当用于不同的上下文环境时，static 关键字具有不同的作用。

在函数定义时，或者代码块之外的变量声明时， static 关键字用于修改标识符的链接属性。从external 改为 internal，即从全局变量改为局部变量。

在代码块内部时，static 关键字用于修改变量的存储类型，从自动变量修改为静态变量。

# <center>作用域、链接属性和存储类型总结</center>

|变量类型|声明位置|是否存储于堆栈|作用域|如果声明为static|
|--|--|--|--|--|--|
|全局|所有代码块之外|否|从声明出到文件尾部|全局变量改为局部变量，不允许从其他源文件访问|
|局部|代码块起始处|是|整个代码块|变量不存储在堆栈中，从自动变量改为静态变量|
|形参|函数头部|是|整个函数|不允许|
